The goal is to write a function to count the frequency of different characters in a given string. For example: character_frequency('google') should return a `dict`: {'g': 2, 'o': 2, 'l': 1, 'e': 1}.

```python
def character_frequency(s):
    ret = dict()
    for i in s:
      if i in ret:
        ret[i] = ret[i] + 1
      else:
        ret[i] = 1
    #ret['0'] = 0
    #ret['1'] = 1
    return ret
```
